```python
path_data_dir = Path.cwd().parent.parent / "data"
path_processed_dir1 = Path(Path.cwd().parent.parent / 'data/processed/cnc/cnc_features_comp')
print(path_processed_dir1)

df1 = pd.read_csv(
    path_processed_dir1 / "cnc_features_54_comp.csv",
)

path_processed_dir2 = Path(Path.cwd().parent.parent / 'data/processed/cnc/cnc_features_comp_extra')

df2 = pd.read_csv(
    path_processed_dir2 / "cnc_features_54_comp_extra.csv",
)
```

```python
df1 = load_cnc_features(
    path_data_dir, path_processed_dir1, "cnc_features_54_comp.csv", "high_level_labels_MASTER_update2022-08-18_with_case.csv"
)
df1 = df1.sort_values(by=['id']).reset_index(drop=True)
print(df1.shape)
df1.head()
```
